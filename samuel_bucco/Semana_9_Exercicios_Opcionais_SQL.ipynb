## Exercício 1 — Cursos mais caros

Liste todos os cursos com preço acima de R$ 200, mostrando:

- `nome_curso`
- `categoria`
- `carga_horaria`
- `preco`

Ordene do maior preço para o menor.


```python
# Escreva sua consulta SQL aqui
query = """
SELECT
	nome_curso,
    categoria,
    carga_horaria,
    preco
FROM cursos
WHERE preco > 200
ORDER BY preco DESC    
"""

con.execute(query).df()
```

## Exercício 2 — Matrículas por status

Conte quantas matrículas existem em cada status:

- `Concluído`
- `Em andamento`
- `Cancelado`

Ordene do maior volume para o menor.


```python
query = """
SELECT
	status,
    COUNT(matricula_id) AS matriculas
FROM matriculas
GROUP BY status
ORDER BY status DESC
"""

con.execute(query).df()
```

## Exercício 3 — Alunos e cursos matriculados

Use `INNER JOIN` para listar as matrículas mostrando:

- nome do aluno
- estado do aluno
- nome do curso
- categoria do curso
- status da matrícula

Ordene pelo nome do aluno.


```python
query = """
SELECT
	a.nome_aluno AS aluno,
    a.estado,
    c.nome_curso AS curso,
    c.categoria AS categoria_curso,
    m.status
FROM matriculas m
JOIN cursos c
	ON m.curso_id = c.curso_id
JOIN alunos a
	ON a.aluno_id = m.aluno_id
"""

con.execute(query).df()
```

## Exercício 4 — Receita por categoria de curso

Calcule a receita potencial por categoria, considerando o preço do curso em cada matrícula.

Mostre:

- categoria
- quantidade de matrículas
- receita total
- preço médio dos cursos matriculados

Ordene pela maior receita total.


```python
query = """
SELECT
	c.categoria,
    COUNT(m.matricula_id) AS qtd_matriculas,
    SUM(c.preco) AS receita_total,
    AVG(c.preco) AS preco_medio
FROM matriculas m
JOIN cursos c
	ON m.curso_id = c.curso_id
GROUP BY c.categoria
ORDER BY receita_total DESC
"""

con.execute(query).df().round(2)
```

## Exercício 5 — Categorias com receita acima de R$ 500

Usando `GROUP BY` e `HAVING`, liste apenas as categorias de curso com receita total acima de R$ 500.

Mostre:

- categoria
- receita total

Ordene pela receita total em ordem decrescente.


```python
query = """
SELECT
	c.categoria,
	-- COUNT(m.curso_id) AS qtd_cursos,
    SUM(c.preco) AS receita_total
FROM matriculas m
JOIN cursos c
	ON c.curso_id = m.curso_id
GROUP BY c.categoria
HAVING receita_total > 500
"""

con.execute(query).df().round(2)
```

## Exercício 6 — Minutos assistidos por aluno

Calcule o total de minutos assistidos por aluno.

Use as tabelas:

- `alunos`
- `matriculas`
- `aulas_assistidas`

Mostre:

- nome do aluno
- estado
- total de minutos assistidos

Ordene do maior total para o menor.


```python
query = """
SELECT
	a.nome_aluno,
    a.estado,
    -- COUNT(m.matricula_id) AS qtd_matriculas,
    SUM(aa.minutos_assistidos) AS total_minutos_assistidos
FROM matriculas m
JOIN alunos a
	ON m.aluno_id = a.aluno_id
JOIN aulas_assistidas aa
	ON m.matricula_id = aa.matricula_id
GROUP BY a.nome_aluno, a.estado
ORDER BY total_minutos_assistidos DESC
"""

con.execute(query).df()
```

## Exercício 7 — Matrículas por mês

Use `DATE_TRUNC` para calcular a quantidade de matrículas por mês.

Mostre:

- mês de referência
- quantidade de matrículas

Ordene o resultado em ordem cronológica.


```python
query = """
SELECT
	DATE_TRUNC('month', CAST(m.data_matricula AS DATE)) AS mes_referencia,
    COUNT(m.matricula_id) AS qtd_matriculas
FROM matriculas m
GROUP BY mes_referencia
ORDER BY mes_referencia
"""

con.execute(query).df()
```

## Exercício 9 — Notas com razão social do cliente

Use `INNER JOIN` para listar as notas fiscais com o nome do cliente.

- `razao_social`
- `segmento`
- `uf_destino`

## Exercício 10 — Valor total vendido por UF de destino

Calcule o valor total dos itens das notas por `uf_destino`.

Considere apenas notas com status `Autorizada`.

Mostre:

- UF de destino
- quantidade de notas distintas
- valor total dos itens
- valor total de imposto

Ordene pelo maior valor total dos itens.


```python
query = """
SELECT
	nf.uf_destino,
    COUNT(DISTINCT inf.nota_id) AS qtd_nf_unicas,
    SUM(inf.valor_item) AS valor_total_itens,
    SUM(inf.valor_imposto) AS valor_total_imposto
FROM itens_notas inf
JOIN notas_fiscais nf
	ON inf.nota_id = nf.nota_id
GROUP BY nf.uf_destino
ORDER BY valor_total_itens DESC
"""

con.execute(query).df()
```

## Exercício 11 — Receita fiscal por tipo: Produto ou Serviço

Use as tabelas `notas_fiscais`, `itens_notas` e `produtos_fiscais` para calcular o valor total por tipo.

Mostre:

- tipo
- quantidade de itens
- valor total
- imposto total

Considere apenas notas autorizadas.


```python
query = """
SELECT
	pf.tipo,
    COUNT(inf.quantidade) AS qtd_itens,
    SUM(inf.valor_item) AS valor_total,
    SUM(inf.valor_imposto) AS imposto_total
FROM itens_notas inf
JOIN produtos_fiscais pf
	ON inf.produto_id = pf.produto_id
JOIN notas_fiscais nf
	on inf.nota_id = nf.nota_id
WHERE nf.status_nota = 'Autorizada'
GROUP BY pf.tipo

"""

con.execute(query).df()
```

## Exercício 12 — Clientes com faturamento acima de R$ 5.000

Usando `GROUP BY` e `HAVING`, liste os clientes com valor total autorizado acima de R$ 5.000.

Mostre:

- razão social
- segmento
- valor total
- imposto total

Ordene pelo maior valor total.


```python
query = """
SELECT
	cf.razao_social,
    cf.segmento,
    SUM(inf.valor_item) AS valor_total,
    SUM(inf.valor_imposto) AS imposto_total
FROM itens_notas inf
JOIN notas_fiscais nf
	ON nf.nota_id = inf.nota_id
JOIN clientes_fiscais cf
	ON cf.cliente_id = nf.cliente_id
WHERE nf.status_nota = 'Autorizada'
GROUP BY cf.razao_social, cf.segmento
HAVING valor_total > 5000
ORDER BY valor_total DESC
"""

con.execute(query).df()
```

## Exercício 13 — Emissões por trimestre

Use `EXTRACT(QUARTER FROM data_emissao)` para contar quantas notas foram emitidas por trimestre.

Mostre:

- ano
- trimestre
- quantidade de notas

Ordene por ano e trimestre.


```python
query = """
SELECT
    EXTRACT(YEAR FROM CAST(nf.data_emissao AS DATE)) AS ano,
	EXTRACT(QUARTER FROM CAST(nf.data_emissao AS DATE)) AS trimestre,
    COUNT(nf.nota_id) AS qtd_notas_fiscais_emitidas
FROM notas_fiscais nf
GROUP BY 1, 2
ORDER BY 1, 2
"""

con.execute(query).df()
```

## Exercício 18 — Fluxo financeiro mensal

Usando a tabela `movimentacoes_bancarias`, calcule entrada, saída e saldo por mês.

Dica: use `CASE WHEN` dentro do `SUM`.

Mostre:

- mês de referência
- total de entradas
- total de saídas
- saldo do mês

Ordene cronologicamente.


```python
query = """
SELECT
	EXTRACT(YEAR FROM CAST(data_movimento AS DATE)) AS ano_referencia,
	EXTRACT(MONTH FROM CAST(data_movimento AS DATE)) AS mes_refencia,
    SUM(
		CASE
			WHEN tipo_movimento = 'Entrada' 
            THEN valor
            ELSE 0
        END
    ) AS total_entradas,
    SUM(
		CASE
			WHEN tipo_movimento = 'Saída' 
            THEN valor
            ELSE 0
        END
    ) AS total_saidas,
    SUM(
		CASE
			WHEN tipo_movimento = 'Entrada'
            THEN valor
            ELSE -valor
        END
    ) AS saldo_mes
FROM movimentacoes_bancarias
GROUP BY 1, 2
ORDER BY 1, 2   
"""

con.execute(query).df()
```

## Exercício 19 — Atraso em contas a pagar

Liste as contas pagas depois do vencimento.

Mostre:

- fornecedor
- data de vencimento
- data de pagamento
- valor
- dias de atraso

Dica: no DuckDB, você pode subtrair datas para calcular diferença em dias.


```python
query = """
SELECT
	fornecedor,
    data_vencimento,
    data_pagamento,
    valor,
    DATEDIFF('day', data_vencimento, data_pagamento) AS dias_atraso
FROM contas_pagar
WHERE status = 'Pago'
GROUP BY 1, 2, 3 , 4
HAVING dias_atraso > 0
"""

con.execute(query).df()
```

## Exercício 20 — Análise financeira integrada

Monte uma consulta que responda:

**Qual foi o total de entradas, saídas e saldo acumulado considerando todas as movimentações bancárias?**

Mostre apenas uma linha com:

- total de entradas
- total de saídas
- saldo final


```python
query = """
SELECT
	SUM(
		CASE
			WHEN tipo_movimento = 'Entrada'
			THEN valor
        END
	) AS total_entradas,
    SUM(
		CASE
			WHEN tipo_movimento = 'Saída'
            THEN valor
        END
    ) AS total_saidas,
    SUM(
		CASE
			WHEN tipo_movimento = 'Entrada' THEN valor
            WHEN tipo_movimento = 'Saída' THEN -valor
        END
    ) AS saldo_final
FROM movimentacoes_bancarias
"""

con.execute(query).df()
```

```python
query = """
SELECT
	c.categoria,
    SUM(c.preco) AS receita_potencial
FROM matriculas m
JOIN cursos c
	ON c.curso_id = m.curso_id
GROUP BY c.categoria
ORDER BY receita_potencial DESC
LIMIT 1
"""

con.execute(query).df()
```

```python
query = """
SELECT
	nf.uf_destino AS estado,
    SUM(inf.valor_item) AS valor_autorizado_nf
FROM itens_notas inf
JOIN notas_fiscais nf
	ON inf.nota_id = nf.nota_id
WHERE nf.status_nota = 'Autorizada'
GROUP BY 1
ORDER BY 2 DESC
LIMIT 1
"""

con.execute(query).df()
```

```python
query = """
SELECT
	cc.centro_custo,
    SUM(cp.valor) AS despesa_total
FROM contas_pagar cp
JOIN centros_custo cc
	ON cp.centro_custo_id = cc.centro_custo_id
GROUP BY cc.centro_custo
ORDER BY despesa_total DESC
LIMIT 1
"""

con.execute(query).df()
```

```python
query = """
SELECT
	EXTRACT(YEAR FROM data_movimento) AS ano,
	EXTRACT(MONTH FROM data_movimento) AS mes,
    SUM(
		CASE
			WHEN tipo_movimento = 'Entrada' THEN valor
            WHEN tipo_movimento = 'Saída' THEN -valor
            ELSE 0
        END
    ) AS saldo
FROM movimentacoes_bancarias
GROUP BY 1, 2
ORDER BY 3 DESC
LIMIT 1
"""

con.execute(query).df()
```
